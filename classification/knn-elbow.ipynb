Sample code for choosing value of $k$ for $k$-NN

Code is based-on https://www.geeksforgeeks.org/ml-determine-the-optimal-value-of-k-in-k-means-clustering/

```python
from sklearn.datasets.samples_generator import make_blobs 

X, y = make_blobs(n_samples=100, centers=4, cluster_std=1.0, n_features = 2) 
```

```python
def evaluate_knn(X, K, ax=None):
    from sklearn.cluster import KMeans 
    
    k_inertia = {}

    for k in range(1, K + 1): 
        kmeanModel = KMeans(n_clusters=k)
        kmeanModel.fit(X) 
        # Inertia is the sum of squared distances to the centers for each point
        k_inertia[k] = kmeanModel.inertia_
        
    if ax:
        ax.plot(list(k_inertia.keys()), list(k_inertia.values()), 'o-') 
        ax.set_ylabel("Inertia")
        ax.set_xlabel("k")
    
    return k_inertia
```

# Toy data

```python
from sklearn.datasets.samples_generator import make_blobs 
import matplotlib.pyplot as plt  

%matplotlib inline

X, y = make_blobs(n_samples=100, centers=4, cluster_std=1.0, n_features = 2) 
x1, x2 = X[:, 0], X[:, 1]
fig, axes = plt.subplots(1, 2, figsize=(12, 4))

axes[0].scatter(x1, x2)
k_inertia = evaluate_knn(X, K=8, ax=axes[1])
```
